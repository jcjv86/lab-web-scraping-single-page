```python
def scrape_hot100():
    url = "https://www.billboard.com/charts/hot-100/"
    response = requests.get(url)
    print('Status Response Code - if 200 all is good!\n',response.status_code)
    soup = BeautifulSoup(response.text, 'html.parser')
    top = soup.select('#post-1479786 > div.pmc-paywall > div > div > div > div.chart-results-list.\/\/.lrv-u-padding-t-150.lrv-u-padding-t-050\@mobile-max')
    
    songs_raw = soup.select('#title-of-a-story')
    songs_total = [i.get_text() for i in songs_raw]
    songs_filtered = songs_total[6::4]
    songs_filtered = songs_filtered[:-4]
    songs=[i.rstrip().lstrip() for i in songs_filtered]
    
    artist_raw = soup.select('.c-label')
    artist_total = [i.get_text() for i in artist_raw]
    artists = [i.rstrip().lstrip() for i in artist_total]

    artist_index = []
    for index in range(0, len(artists)-1):
        try:
            int(artists[index])
            
        except:
            artist_index.append(index)
    
    artist_raw2 = [artists[i] for i in artist_index]
    artist = []
    for i in artist_raw2:
        if i != 'NEW':
            if i!= '-':
                if i!= 'RE-\nENTRY':
                    artist.append(i)
    artist = artist[:100]
    numbers = [i for i in range(1,101)]
    
    top_100 = pd.DataFrame(data = list(zip(numbers, song, artist)), columns=['Number', 'Song', 'Artist'])
    display(top_100.style.hide(axis='index'))
    return top_100.to_csv('top_100.csv', index = False)    
```
